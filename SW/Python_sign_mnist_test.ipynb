```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import datasets, layers, models, utils
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

test_df = pd.read_csv('/content/drive/MyDrive/공부/아이덱/수화/test.csv')
print(test_df.shape)

test_images=test_df.values[0:,1:]
test_labels = test_df.values[0:,0]
test_labels = utils.to_categorical(test_labels)

print(test_images.shape,test_labels.shape)
#normalizing the dataset
test_images=test_images/255
test_images=test_images.reshape((20,28,28,1))
print(test_images.shape,test_labels.shape)
```

```python
sign_mnist = models.load_model('/content/drive/MyDrive/공부/아이덱/수화/sign_mnist.h5')
sign_mnist.summary()
```

```python

for n in range(20):
  test_images_n = test_images[n]
  flog = open("data.log", "w")
  test_labels_n = np.argmax(test_labels[n])
  for i in range(28):
      for j in range(28):
          print("%d "%abs(test_images_n[i][j][0]*255), file = flog, end=' ')
          print("", file = flog) 
  print("", file = flog) 
  for i in range(28):
      for j in range(28):
          print("8'd%d, "%abs((test_images_n[i][j][0]*255).astype(int)), file = flog, end=' ') 
      print("", file = flog)  
  print("", file = flog)     
  print(test_labels[n], file = flog)
  test_images_n = test_images_n.reshape(1,28,28,1)
  flog.close()
  predicted_result_2 = sign_mnist.predict(test_images_n)
  predicted_labels_2 = np.argmax(predicted_result_2, axis=1)
  print(predicted_labels_2)
```
